```python
import pandas as pd

bank_df = pd.read_csv('bank-prep.csv', sep=',')
bank_df.head()
```

```python
print(bank_df.shape)
print(bank_df.dtypes)
```

```python
# 스케일링
from sklearn.preprocessing import MinMaxScaler
bank_df = bank_df.drop('y', axis=1)

mc = MinMaxScaler()
mc.fit(bank_df)

bank_df_mc = pd.DataFrame(mc.transform(bank_df), columns=bank_df.columns)
bank_df_mc.head()
```

```python
# 스케일링 2
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(bank_df)

bank_df_sc = pd.DataFrame(sc.transform(bank_df), columns=bank_df.columns)
bank_df_sc.head()
```

```python
# 각 변수의 평균과 표준편차 값 확인
print(bank_df_sc.mean())
print(bank_df_sc.std())
```

고객의 그룹화

계층형 클러스터링, 비계층형 클러스터링

유클리드 거리 / 맨하튼 거리 / 코사인 유사도


```python
# 계층형 클러스터링 : 거리가 가까운 데이터부터 순서대로 병합해 그룹형성 / 결과는 덴드로그램으로 표현
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

hcls = linkage(bank_df_sc, metric='euclidean', method='ward') # 데이터의 병합 워드방법(중심과 각 데이터 간 거리의 제곱 합)
dendrogram(hcls)

plt.ylabel('dist')
plt.show()
```

```python
# 고객별 그룹확인
from scipy.cluster.hierarchy import fcluster

cst_group = fcluster(hcls, 100, criterion='distance') # 데이터 사리의 거리를 유클리드로 측정하고, 거리의 경계값을 100으로
print(cst_group)
```

```python
# 비계층형 클러스터링
# k-means 법
from sklearn.cluster import KMeans
kcls = KMeans(n_clusters=10)
cst_group = kcls.fit_predict(bank_df_sc)
print(cst_group)
```

```python
# 클러스터 ID 별 데이터 분포 확인
for i in range(10):
  labels = bank_df_sc[cst_group == i]
  plt.scatter(labels['age'], labels['balance'], label=i)

plt.legend()
plt.xlabel('age')
plt.ylabel('balance')
plt.show()
```

```python
# 결과 결합
bank_df_sc['group'] = cst_group
bank_df_sc.head()
```

```python
print(bank_df_sc['group'].value_counts())
```

```python
# 각 그룹의 통계량 계산 및 성질 파악
bank_df_sc[bank_df_sc['group']==0].describe()
```

잠재적인 요구 추출

```python
# 주성분 분석
from sklearn.decomposition import PCA
pca = PCA(0.80) # 누적 기여율이 80% 도달할 때 까지 주성분을 추출
bank_df_pca = pca.fit_transform(bank_df_sc)

print(pca.n_components_)
print(bank_df_pca.shape)
```

```python
# 목적변수별 분포 확인
y = pd.read_csv('bank-prep.csv', sep=',')['y']
bank_df_pca = pd.DataFrame(bank_df_pca)
bank_df_pca['y'] = y
print(bank_df_pca.shape)
```

```python
# 목적변수별 산포도 확인
import matplotlib.pyplot as plt
%matplotlib inline

bank_df_pca_0 = bank_df_pca[bank_df_pca['y'] == 0]
bank_df_pca_0 = bank_df_pca_0.drop('y', axis=1)
plt.scatter(bank_df_pca_0[0], bank_df_pca_0[1], c='red', label=0)

bank_df_pca_1 = bank_df_pca[bank_df_pca['y'] == 1]
bank_df_pca_1 = bank_df_pca_1.drop('y', axis=1)
plt.scatter(bank_df_pca_1[0], bank_df_pca_1[1], c='blue', label=1)

plt.legend()
plt.xlabel('1st-comp')
plt.ylabel('2nd-comp')
plt.show()
```
